```python
#  Import Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
```

```python
# ðŸŽ¨ Plot Settings
sns.set_style("whitegrid") 
sns.set_palette("Set2")     
pd.set_option("display.float_format", lambda x: "%.2f" % x)  # Format 
```

```python
df = pd.read_csv("C:/Users/tusha/Downloads/train.csv")

print("Dataset Shape:", df.shape)

```

```python
# ===============================
# 2. Missing Values Analysis
# ===============================
missing_percent = (df.isnull().sum()/len(df))*100
missing_data = pd.DataFrame({"MissingCount": df.isnull().sum(),
                             "MissingPercent": missing_percent})
missing_data = missing_data[missing_data["MissingCount"] > 0].sort_values("MissingPercent", ascending=False)
display(missing_data.head(20))
```

```python
plt.figure(figsize=(10,6))
missing_data["MissingPercent"].plot(kind="bar")
plt.title("Top Missing Features (%)")
plt.ylabel("Percent Missing")
plt.show()
```

```python
# ===============================
# 3. Duplicate Check
# ===============================
print("Duplicate rows:", df.duplicated().sum())
```

```python
# ===============================
# 4. Outlier Detection
# ===============================
plt.figure(figsize=(8,5))
sns.boxplot(x=df["SalePrice"])
plt.title("Boxplot of SalePrice")
plt.show()


outliers = df[(df['GrLivArea'] > 4000) & (df['SalePrice'] < 300000)]
print("Potential Outliers (GrLivArea vs SalePrice):")
display(outliers[["Id","GrLivArea","SalePrice","Neighborhood"]])
```

```python
# ===============================
# 5. Univariate Analysis
# ===============================
# Numerical variables
num_vars = df.select_dtypes(include=[np.number]).columns
summary = pd.DataFrame({
    "Mean": df[num_vars].mean(),
    "Median": df[num_vars].median(),
    "TrimmedMean": df[num_vars].apply(lambda x: stats.trim_mean(x, 0.1)),
    "Variance": df[num_vars].var(),
    "StdDev": df[num_vars].std(),
    "Range": df[num_vars].max() - df[num_vars].min()
})
display(summary.head())
```

```python
# Distribution of SalePrice
plt.figure(figsize=(10,6))
sns.histplot(df["SalePrice"], kde=True)
plt.title("Distribution of SalePrice")
plt.show()
```

```python
# Log transformation
plt.figure(figsize=(10,6))
sns.histplot(np.log1p(df["SalePrice"]), kde=True)
plt.title("Log-Transformed SalePrice")
plt.show()
```

```python
# Categorical variables
cat_vars = df.select_dtypes(include=["object"]).columns
for col in cat_vars[:3]:  # show first 3 as example
    print(f"\n{col} value counts:")
    print(df[col].value_counts().head())
    print("Mode:", df[col].mode()[0])

    plt.figure(figsize=(8,4))
    sns.countplot(y=col, data=df, order=df[col].value_counts().index)
    plt.title(f"{col} Frequency")
    plt.show()
```

```python
# ===============================
# 6. Bivariate Analysis
# ===============================
# Scatter: GrLivArea vs SalePrice
plt.figure(figsize=(8,5))
sns.scatterplot(x="GrLivArea", y="SalePrice", data=df, alpha=0.6)
plt.title("GrLivArea vs SalePrice")
plt.show()
```

```python
# Correlation Matrix
corr = df.corr(numeric_only=True)
plt.figure(figsize=(12,8))
sns.heatmap(corr, cmap="coolwarm", center=0)
plt.title("Correlation Heatmap")
plt.show()
```

```python
# Top correlations with SalePrice
top_corr = corr["SalePrice"].abs().sort_values(ascending=False).head(15)
plt.figure(figsize=(8,6))
sns.barplot(x=top_corr.values, y=top_corr.index)
plt.title("Top 15 Correlations with SalePrice")
plt.show()
```

```python
# Boxplot: Neighborhood vs SalePrice
plt.figure(figsize=(14,6))
neigh_order = df.groupby("Neighborhood")["SalePrice"].median().sort_values().index
sns.boxplot(x="Neighborhood", y="SalePrice", data=df, order=neigh_order)
plt.xticks(rotation=90)
plt.title("SalePrice by Neighborhood")
plt.show()
```

```python
# ===============================
# 7. Dataset-Specific EDA
# ===============================

# Year Built vs SalePrice (Median)
# Year Built vs SalePrice (Median)
plt.figure(figsize=(12,6))
sns.lineplot(x="YearBuilt", y="SalePrice", data=df, estimator=np.median, errorbar=None)
plt.title("Median SalePrice by Year Built")
plt.xlabel("Year Built")
plt.ylabel("SalePrice")
plt.tight_layout()
plt.show()
```

```python
# OverallQual vs SalePrice
plt.figure(figsize=(10,6))
sns.boxplot(x="OverallQual", y="SalePrice", data=df)
plt.title("SalePrice by Overall Quality")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(x="GarageCars", y="SalePrice", data=df)
plt.title("SalePrice by Garage Capacity")
plt.show()
```

```python
# Basement area
plt.figure(figsize=(8,5))
sns.scatterplot(x="TotalBsmtSF", y="SalePrice", data=df)
plt.title("Total Basement Area vs SalePrice")
plt.show()
```

```python
# Living area vs Basement
plt.figure(figsize=(8,5))
sns.scatterplot(x="GrLivArea", y="TotalBsmtSF", data=df, alpha=0.6)
plt.title("GrLivArea vs TotalBsmtSF")
plt.show()
```

```python
# Neighborhood premium vs low-price
neighborhood_prices = df.groupby("Neighborhood")["SalePrice"].median().sort_values(ascending=False)
print(" Premium Neighborhoods:\n", neighborhood_prices.head(5))
print("\n Lower-Price Neighborhoods:\n", neighborhood_prices.tail(5))
```

```python
# Select numeric features and compute correlations
num_df = df.select_dtypes(include=[np.number])
corr_all = num_df.corr()['SalePrice'].drop('SalePrice').sort_values(ascending=False)

# Top k correlated features
top_k = 12
top_corr = corr_all.head(top_k)
display(top_corr.to_frame('corr_with_SalePrice'))

# ðŸ“Š Barplot of top correlations
plt.figure(figsize=(8,5))
sns.barplot(x=top_corr.values, y=top_corr.index, hue=top_corr.index, palette='coolwarm', legend=False)
plt.title('Top correlations with SalePrice')
plt.xlabel('Pearson correlation')
plt.ylabel('Feature')
plt.tight_layout()
plt.show()

# ðŸ”¥ Focused heatmap for top features + SalePrice
sel_feats = ['SalePrice'] + top_corr.index.tolist()
plt.figure(figsize=(8,6))
sns.heatmap(num_df[sel_feats].corr(), annot=True, fmt='.2f', cmap='coolwarm', square=True)
plt.title('Heatmap: SalePrice and top-correlated features')
plt.tight_layout()
plt.show()
```

```python
# SaleCondition
plt.figure(figsize=(10,5))
sns.boxplot(x="SaleCondition", y="SalePrice", data=df)
plt.title("SalePrice by Sale Condition")
plt.show()
```

```python
# Basement & Garage Features Correlation
features = ["TotalBsmtSF","GarageArea","GarageCars","1stFlrSF","GrLivArea","SalePrice"]
plt.figure(figsize=(8,6))
sns.heatmap(df[features].corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Basement/Garage Features Correlation")
plt.show()
```

```python
# Extreme price outliers
outliers_ext = df[(df['SalePrice'] > 700000) | (df['SalePrice'] < 50000)]
print("Extreme Price Outliers:")
display(outliers_ext[["Id","SalePrice","GrLivArea","Neighborhood","OverallQual"]])
```
